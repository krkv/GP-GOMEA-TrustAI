```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
np.random.seed(0)
```

```python
df = pd.read_csv('bikes.csv', index_col=0)
df.head()
```

```python
df.describe().T
```

```python
df2 = df.loc[:,('season',
                'yr',
                'mnth',
                'holiday',
                'weekday',
                'workingday',
                'weathersit',
                'temp',
                'atemp',
                'hum',
                'windspeed',
                'cnt')]

df2.rename(columns={
                    'yr': 'year',
                    'mnth': 'month',
                    'weekday': 'week_day',
                    'workingday': 'working_day',
                    'weathersit': 'weather',
                    'temp': 'temperature',
                    'atemp': 'feeling_temperature',
                    'hum': 'humidity',
                    'windspeed': 'wind_speed',
                    'cnt': 'y'
                  }, inplace=True)
df2.head()
```

```python
X = df2
y = X.pop('y')
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
y_train.dtypes
```

```python
X_train_values = X_train.values.astype(float)
y2_train_values = y_train.values.astype(float)
print(X_train_values.dtype, y2_train_values.dtype)
```

```python
from pyGPGOMEA import GPGOMEARegressor as GPG
```

```python
ea = GPG(gomea=True, silent=False)
ea.fit(X_train_values, y2_train_values)
```

```python
model = ea.get_model().replace("p/","/").replace("plog","log")
```

```python
import sympy
model = str(sympy.simplify(model))
```

```python
from sklearn.metrics import mean_squared_error

if model == "0":
	print("Warning: sympy couldn't make it due to numerical precision")
	# re-set to non-simplified model
	model = ea.get_model().replace("p/","/").replace("plog","log")
print('Model found:', model)
print('Evaluations taken:', ea.get_evaluations()) # care: this is not correct if multiple threads were used when fitting
print('Test RMSE:', np.sqrt( mean_squared_error(y_test, ea.predict(X_test.values)) ))
```

```python
import pickle as pkl

with open("bikes_gp_gomea.pkl", "wb") as f:
    pkl.dump(ea, f)
```
